```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
BASE_PATH = "/content/drive/MyDrive/Change Detection/Output Images/"
```

```python
imgA = cv2.imread(os.path.join(BASE_PATH, "A.png"))
imgB = cv2.imread(os.path.join(BASE_PATH, "B.png"))
gt = cv2.imread(os.path.join(BASE_PATH, "GT.png"), 0)

No_GT_Predicted = cv2.imread(os.path.join(BASE_PATH, "No_GT_Predicted.png"), 0)
Some_GT_Predicted = cv2.imread(os.path.join(BASE_PATH, "Some_GT_Predicted.png"), 0)
U_Net_Predicted = cv2.imread(os.path.join(BASE_PATH, "U-Net_Predicted.png"), 0)

imgA = cv2.cvtColor(imgA, cv2.COLOR_BGR2RGB)
imgB = cv2.cvtColor(imgB, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(14, 8))

# Row 1
plt.subplot(2,3,1)
plt.imshow(imgA)
plt.title("A (Before)")
plt.axis("off")

plt.subplot(2,3,2)
plt.imshow(imgB)
plt.title("B (After)")
plt.axis("off")

plt.subplot(2,3,3)
plt.imshow(gt, cmap='gray')
plt.title("Ground Truth")
plt.axis("off")

# Row 2
plt.subplot(2,3,4)
plt.imshow(No_GT_Predicted, cmap='gray')
plt.title("No GT")
plt.axis("off")

plt.subplot(2,3,5)
plt.imshow(Some_GT_Predicted, cmap='gray')
plt.title("Some GT")
plt.axis("off")

plt.subplot(2,3,6)
plt.imshow(U_Net_Predicted, cmap='gray')
plt.title("U-Net")
plt.axis("off")

plt.tight_layout()
plt.savefig("final_comparison.png", dpi=300)
plt.show()
```
